# Introduction 
cloud comptuting : on demand delivery of IT resources over the internet with pay as you go pricing . like server , storage , data base 

Types of cloud comptuting : 
there are three types of cloud comptuting that is 
1: IaaS : infrastructure as a service
2: SaaS : sofware as a service 
3: PaaS : platform as a service

# IAM in AWS stands for Identity and Access Management

It is the enterprise-grade access governance layer of AWS that lets you securely control who can access what, ensuring the right people and systems have the right permissions at the right time.
